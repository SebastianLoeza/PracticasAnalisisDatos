### Ejercicio 4.
Pedir dos numeros al usuario y hacer todas las operaciones básicas de una calculadora por pantalla

```python
num1=int(input("Ingrese el primer numero: "))
num2=int(input("Ingrese el segundo número: "))

print(num1, "+",num2,"=", num1+num2 )
print(num1, "-",num2,"=", num1-num2 )
print(num1, "*",num2,"=", num1*num2 )
print(num1, "/",num2,"=", num1/num2 )
```

### Ejercicio 10.
Tiene que pedir la calificacion de 15 alumnos y sacara por pantalla cuantos han aprobado y cuandos no

```python
aprobado=0
reprobado=0

for i in range(15):
  cal=int(input("Introduce la calificacion del alumno:"))
  if cal>=6:
    aprobado+=1
  else:
    reprobado+=1
print("El número de alumnos aprobados es:",aprobado)
print("El número de alumnos reprobados es:",reprobado)
```
