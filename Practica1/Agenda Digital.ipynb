**Pista 4**: Recuerda que debes escribir una función, te recomiendo que tu función tenga una definición similar a la siguiente:
```

    <sentencias que leen la agenda digital>
```

````python
def leer_agenda():

  agenda_digital_lectura= open("agenda.txt","r")
  agenda_digital = agenda_digital_lectura.readlines()
  agenda_digital_lectura.close()


  data_cleaned = [line.strip() for line in agenda_digital if line.strip()]

  agenda = {}

  current_name = None
  for line in data_cleaned:
      if line.endswith(":"):
          current_name = line[:-1]
          agenda[current_name] = {}
      else:
          key, value = line.split(":", 1)
          key = key.strip()
          value = value.strip()
          agenda[current_name][key] = value
  return (agenda)
````

````python
def buscar_contacto(nombre):
    agenda=leer_agenda()
    if nombre in agenda:
        contacto = agenda[nombre]
        print(f"Información de contacto para {nombre}:")
        print("Alias:"+ contacto["Alias"])
        print("Teléfono:"+ contacto["Telefono"])
        print("Dirección:"+contacto['Direccion'])
        print("Email:"+contacto['Email'])
    else:
        print(f"{nombre} no se encuentra en la agenda.")


````
