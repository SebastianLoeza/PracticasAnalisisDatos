```python
#FUNCIÓN PARA HACER UNA FUNCIÓN DE MONTECARLO.
def simulacion_montecarlo():
    #Declarar los arreglos de x y y y agregamos 1000 numeros random entre 0 y 1
    x=[]
    y=[]
    for i in range(1000):
        x_uno=random.random()
        x.append(x_uno)
        y_uno= random.random()
        y.append(y_uno)
    x=np.array(x)
    y=np.array(y)
    
    z=[]
    for i in range(len(x)):
        z.append(((x[i])**2)+((y[i])**2))

    count_uno=0
    count_cero=0
    for i in range(len(z)):
        if(z[i]>=1):
            count_uno+=1
        elif (z[i]<1):
            count_cero+=1
    #Probabilidad de que caiga adentro del círculo
    p_cero=count_cero/1000

    return p_cero
```

```python
valor_pi=[]
for i in range(100):
    valor_pi.append(simulacion_montecarlo())
```

```python
valor_pi=np.array(valor_pi)
```

```python
ans=valor_pi.mean()
```
